```python
import openai
from langchain.embeddings.openai import OpenAIEmbeddings
import streamlit as st
import os

openai_api_key = st.secrets["openai_api_key"]

# Create an OpenAIEmbeddings instance
model_name = 'text-embedding-ada-002'
embeddings = OpenAIEmbeddings(model=model_name, openai_api_key=openai_api_key)

# Embed your query
query = "What was Jesus' first miracle?"
query_vector = embeddings.embed_query(query)
```

```python
len(query_vector)
```

```python
import pinecone
from pinecone import Pinecone

# Initialize the Pinecone client
pc = Pinecone(api_key=st.secrets["pinecone_api_key"])

# Connect to your index
index_name = "full-bible-index"
index = pc.Index(index_name)

# Query the index using the query vector
query_response = index.query(
    vector=query_vector,
    top_k=3,
    include_values=True,
    include_metadata=True
)

print(query_response)
```

```python
k = 3
texts = []
for i in range(k):
    text = query_response['matches'][i]['metadata']['text']
    texts.append(text) 

texts_joined = "\n PASSAGE: \n".join(texts)
```
